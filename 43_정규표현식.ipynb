```python
# 문자열 판단하기

import re

re.match('hello', 'hello, world')
```

```python
re.search('hello', 'hello, world')
re.search('world!$', 'hello, world!')
```

```python
## 지정된 문자열이 하나라도 포함되어 있는지 판단하기
re.match('hello|world', 'hello')
```

```python
# 범위 판단하기

print(re.match('[0-9]*', '1234'))
print(re.match('[0-9]+', '1234'))
print(re.match('[0-9]+', '1234'))
```

```python
print(re.match('a*b', 'b'))
print(re.match('a+b', 'b'))
print(re.match('a*b', 'aab'))
print(re.match('a+b', 'aab'))
```

```python
## 문자가 한 개만 있는지 판단하기
# 문자?
# [0-9]?

print(re.match('abc?d', 'abd'))
print(re.match('ab[0-9]?c', 'ab3c'))
print(re.match('ab.d', 'abxd'))
```

```python
## 문자 개수 판단하기
# 문자{개수}
# (문자열){개수}

print(re.match('h{3}', 'hhhello'))
print(re.match('(hello){3}', 'hellohellohelloworld'))
```

```python
# [0-9]{개수}
print(re.match('[0-9]{3}-[0-9]{4}-[0-9]{4}', '010-1000-1000'))
print(re.match('[0-9]{3}-[0-9]{4}-[0-9]{4}', '010-1000-100'))
```

```python
## 숫자와 영문 문자를 조합해서 판단하기
# a-z
# A-Z

print(re.match('[a-zA-Z0-9]+', 'Hello1234'))
print(re.match('[A-Z0-9]+', 'hello'))
```

```python
# 가-힣

re.match('[가-힣]+ [가-힣]+', '홍길동')
```

```python
## 특정 문자 범위에 포함되지 않는지 판단하기

# [^범위]*
# [^범위]+

print(re.match('[^A-Z]+', 'Hello'))
print(re.match('[^A-Z]+', 'hello'))
```

```python
# ^[범위]*
# ^[범위]+

print(re.search('^[A-Z]+', 'Hello'))
```

```python
# [범위]*$
# [범위]+$

re.search('[0-9]*$', 'Hello1234')
```

```python
## 특수문자 판단하기

# \특수문자 

print(re.search('\*+', '1 ** 2'))
print(re.match('[$()a-zA-Z0-9]+', '$(document)'))
```

```python
# \d -> [0-9]
# \D -> [^0-9]
# \w -> [a-zA-Z0-9_]
# \W -> [^a-zA-Z0-9_]

print(re.match('\d+', '1234'))
print(re.match('\\D+', 'Hello'))
print(re.match(r'\w+', 'Hello'))  # [a-zA-Z0-9_]+
print(re.match(r'\W+', '(:)'))
```

```python
## 공백 처리하기

print(re.match(r'[a-zA-Z0-9 ]+', 'Hello 1234'))
print(re.match(r'[a-zA-Z0-9]+', 'Hello 1234'))
print(re.match(r'[a-zA-Z0-9\s]+', 'Hello 1234'))
```

```python
# 그룹 사용하기

# (정규표현식) (정규표현식)
# 매치객체.group(그룹숫자)

m = re.match('([0-9]+) ([0-9]+)', '10 295')
print(m.group(1))
print(m.group(2))
print(m.group(0))
print(m.group())
```

```python
print(m.groups())
```

```python
# (?P<이름>정규표현식)
 
m = re.match('(?P<func>[a-zA-Z0-9_][a-zA-Z0-9_]+)\((?P<arg>\w+)\)', 'print(1234)')
print(m.group('func'))
print(m.group('arg'))
```

```python
## 패턴에 매칭되는 모든 문자열 가져오기

# re.findall('패턴', '문자열')

re.findall('[0-9]+', '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
# 문자열 바꾸기
# re.sub('패턴', '바꾼 문자열', '문자열', 바꿀 횟수)

re.sub('apple|orange', 'fruit', 'apple box orange tree')
```

```python
re.sub('[0-9]+', 'n', '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
# 교체함수(매치객체)
# re.sub('패턴', 교체함수, '문자열', 바꿀 횟수)

def multiple10(m):
    n = int(m.group())
    return str(n * 10)

re.sub('[0-9]+', multiple10, '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
re.sub('[0-9]+', lambda m: str(int(m.group()) * 10), '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
# 연습문제

import re

p = re.compile('^[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+$')
emails = ['python@mail.example.com', 'python+kr@example.com',
          'python-dojang@example.co.ke', 'python_10@example.info',
          'python.dojang@e-xample.com',
          'example.com', 'python@example', 'python@example-com']

for email in emails:
    print(p.match(email) != None, end=' ')
```
