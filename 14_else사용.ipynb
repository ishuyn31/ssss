```python
# else와 들여쓰기
x = 10

if x == 10:
    print("10 입니다.")
else:
    print("x에 들어가는 숫자는")
print("10이 아닙니다.")  # 오류는 아니지만 1번 들여써야 제대로 반환됨
```

```python
# if 조건문에 숫자 지정하기

if 0:
    print("참")
else:
    print("거짓")

if 1:
    print("참")
else:
    print("거짓")

if 0x1f:
    print("참")
else:
    print("거짓")

if 0b1000:
    print("참")
else:
    print("거짓")

if 13.5:
    print("참")
else:
    print("거짓")
```

```python
# 조건문에 문자열 지정하기
if "hello":
    print("참")
else:
    print("거짓")

if " ":
    print("참")
else:
    print("거짓")

if "":
    print("참")
else:
    print("거짓")
```

```python
# 조건식을 여러 개 지정하기

x = 10
y = 20

if x == 10 and y == 20:
    print("참")
else:
    print("거짓")
```

```python
# 연습문제 

written_test = 75
coding_test = True

if written_test >= 80 and coding_test == True:
    print('합격')
else:
    print('불합격')
```
