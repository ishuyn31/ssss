```python
for i in range(5):
    for j in range(5):
        print("j:", j, sep="", end=" ")
    print("i:", i, "\\n", sep="")
```

```python
# 사각형을 별 출력하기
for i in range(5):
    for j in range(5):
        print("*", end="")
    print()
```

```python
for i in range(3):
    for j in range(7):
        print("*", end="")
    print()
```

```python
# 계단식으로 별 출력하기     # 시험!
for i in range(5):
    for j in range(5):
        if j <= i:
            print("*", end="")
    print()
```

```python
# 대각선으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j == i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
# 연습문제
for i in range(5):
    for j in range(5):
        if j < i:
            print(" ", end="")
        else:
            print("*", end="")
    print()
```

```python
# 심사문제
height = int(input())
```

```python
height
```

```python
## 1
for i in range(height):
    for j in range(height-i-1):
        print(" ", end="")
    for k in range(2*i+1):
        print("*", end="")
    print() 
```

```python
##2
for i in range(1, height+1):
    space = height - i
    star = 2 * i - 1
    print(" " * space + "*" * star)
```
