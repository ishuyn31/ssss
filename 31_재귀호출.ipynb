```python
# 재귀호출 사용하기

def hello():
    print('Hello, world!')
    hello()
```

```python
hello()
```

```python
## 재귀호출에 종료 조건 만들기

def hello(count):
    if count == 0:
        return
    
    print('Hello, World!', count)

    count -= 1
    hello(count)
```

```python
hello(5)
```

```python
# 재귀호출로 팩토리얼 구하기

def factorial(n):
    if n == 1:
        return 1
    
    return n * factorial(n - 1)
```

```python
print(factorial(5))
```

```python
# 연습문제

def is_palindrome(word):
    
    if len(word) < 2:
        return True
    if word[0] != word[-1]:
        return False
    return is_palindrome(word[1:-1])

print(is_palindrome('hello'))
print(is_palindrome('level'))
```
