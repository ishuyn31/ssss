```python
# 위치 인수와 리스트 언패킹
def print_numbers(a, b, c):
    print(a)
    print(b)
    print(c)
```

```python
print_numbers(10, 20, 30)
```

```python
# 언패킹 사용하기

x = [10, 20, 30]
print_numbers(*x)
```

```python
print_numbers(*[10, 20, 30])
```

```python
def print_numbers(a, b, c):
    print(a)
    print(b)
    print(c)

print_numbers(10, 20)
```

```python
print_numbers(*[10, 20])
```

```python
## 가변 인수 함수 만들기
# def 함수이름(*매개변수):
#     코드

def print_numbers(*args):
    for arg in args:
        print(arg)
```

```python
print_numbers(10)

print_numbers(10, 20, 30, 40)
```

```python
x = [10]
print_numbers(*x)

y = [10, 20, 30 ,40]
print_numbers(*y)
```

```python
# 키워드 인수 사용하기

def personal_info(name, age, address):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)
```

```python
personal_info('홍길동', 30, '서울시 용산구 이촌동')
```

```python
personal_info(name='홍길동', age=30, address='서울시 용산구 이촌동')
```

```python
personal_info(name='홍길동', address='서울시 용산구 이촌동', age=30)
```

```python
# 키워드 인수와 딕셔너리 언패킹
def perdonal_info(name, age, address):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)
```

```python
x = {
    'name': '홍길동',
    'age': 30,
    'address': '서울시 용산구 이촌동'
    }
personal_info(**x)
# => personal_info(name='홍길동', age=30, address='서울시 용산구 이촌동')
```

```python
# 키워드 인수를 사용하는 가변 인수 함수 만들기

def personal_info(**kwargs):
    # print(kwarg)
    # print(type(kwarg))
    for kw, arg in kwargs.items():
        print(kw, ": ", arg, sep="")
```

```python
personal_info(name='홍길동', age=30)
```

```python
personal_info(name='홍길동', age=30, address='서울시 용산구')
```

```python
x = {
    'name': '홍길동'
}

y = {
    'name': '홍길동',
    'age': 30,
    'address': '서울시 용산구'
    }
```

```python
personal_info(**x)
personal_info(**y)
```

```python
# 매개변수에 초깃값 지정하기

def personal_info(name, age, address='비공개'):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)

personal_info('홍길동', 30)
```

```python
# 연습문제
korean, english, math, science = 100, 86, 81, 91

def get_max_score(*args):
    return max(args)

max_score = get_max_score(korean, english, math, science)
print('높은 점수: ', max_score)

max_score = get_max_score(english, science)
print('높은 점수: ', max_score)
```
