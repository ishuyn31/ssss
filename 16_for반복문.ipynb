```python
# for 변수 in range(횟수)
#     반복할 코드

for i in range(10):
    print(f"{i} Hello World!")
```

```python
for _ in range(10):
    print("Hello World!")
```

```python
# 시작하는 숫자와 끝나는 숫자 지정하기
for i in range(5, 12):
    print("hello world", i)
```

```python
for a in range(0, 10, 2):
    print("hello world", a)
```

```python
# 숫자를 감소시키기
for j in range(10, 0, -1):
    print("hello world", j)
```

```python
list(reversed(range(10)))
```

```python
for b in reversed(range(10)):
    print("hello", b)
```

```python
# 입력한 횟수대로 반복하기
count = int(input("반복할 횟수를 입력하시오."))

for i in range(count):
    print("hello world", i)
```

```python
# 시퀀스 객체로 반복하기
a = [10, 20, 30, 40, 50]
for i in a:
    print(i)
    
```

```python
## 튜플
fruits = (
    'apple', 'orange', 'grape'
)

for fruit in fruits:
    print(fruit)
```

```python
for letter in "python":
    print(letter, end=" ")
```

```python
list(reversed("python"))
```

```python
for letter in reversed("python"):
    print(letter)
```

```python
# 연습문제
x = [49, -17, 25, 102, 8, 62, 21]

for i in x:
    print(i * 10, end=' ')
```

```python
# 심사문제
dan = int(input("단을 입력하세요(2~9): "))

for num in range(1, 10):
    print(f"{dan} * {num} = {dan * num}")
```
