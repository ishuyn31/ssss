```python
# 문자열 서식 지정자와 포매팅 사용하기

## 서식 지정자로 문자열 넣기
# "%s" % "문자열"

"I am %s" % "james"
```

```python
name = "maria"
"I am %s" % name
```

```python
## 서식 지정자로 숫자 넣기
# "%d" % 숫자

"I am %d years old" % 20
```

```python
## 서식 지정자로 소수점 표현하기
# "%f" % 숫자

"%f" % 2.3
```

```python
# "%.자리수f" % 숫자
"%.2f" % 2.3
```

```python
"%10.3f" % 2.3
```

```python
## 서식 지정자로 문자열 정렬하기
print("%10s" % "python")
print("%-10s" % "python")
```

```python
## 서식 지정자로 문자열 안에 값 여러 개 넣기
"Today is %d %s" % (3, "April")
```

```python
## format 매서드 사용하기
"hello {0}".format("world")
```

```python
"Hello {0} {1} {2}".format("Python", "Script", 3.6)
```

```python
"{0} {0} {1} {1}".format("Python", "Script")
```

```python
"Hello {language} {version}".format(language="Python", version=3.6)
```

```python
## format 매서드로 문자열 정렬하기
"{0:<10}".format("python")
```

```python
"{0:>10}".format("python")
```

```python
## 숫자 개수 맞추기
"%03d" % 1

"{0:03d}".format(35)
```

```python
"%08.2f" % 3.6
```

```python
"{0:08.2f}".format(3.6)
```
