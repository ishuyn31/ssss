```python
students = {
    "김철수": [90, 80, 70],
    "이영희": [100, 95, 90],
    "박민수": [60, 70, 65]
}

print("===== 학생 성적 결과 =====")

total_avg_sum = 0
student_count = len(students)

for name, score in students.items():
    avg = sum(score) / len(score)
    total_avg_sum += avg
    
    print(f"이름: {name}")
    print(f"점수: {score}")
    print(f"평균: {avg:.1f}\n")

total_avg = total_avg_sum / student_count

print("===== 전체 결과 =====")
print(f"전체 학생 수: {student_count}명")
print(f"전체 평균: {total_avg:.1f}")
```

```python

student_list = []

print("학생 5명의 정보를 차례대로 입력해주세요.\n")

for i in range(5):
    print(f"[{i+1}번째 학생]")
    name = input("이름을 입력하세요: ")
    age = input("나이를 입력하세요: ")
    grade = input("학년을 입력하세요: ")
    
    student_dict = {
        "이름": name,
        "나이": age,
        "학년": grade
    }
    
    student_list.append(student_dict)
    print("-" * 25)

print("\n===== 전체 학생 정보 =====")

for student in student_list:
    print(student)
```

```python
height = int(input("역삼각형의 높이를 입력하세요: "))

for i in range(height):
    for j in range(i):
        print(" ", end="")
    
    for k in range(2 * (height - i) - 1):
        print("*", end="")
        
    print()
```

```python

col, row = map(int, input("가로(col)와 세로(row)를 입력하세요: ").split())

game_map = []
for i in range(row):
    game_map.append(list(input()))

move_y = [-1, -1, -1, 0, 0, 1, 1, 1]
move_x = [-1, 0, 1, -1, 1, -1, 0, 1]

for i in range(row):
    for j in range(col):
        if game_map[i][j] == "*":
            print("*", end="")
        else:
            bomb_count = 0 
            
            for k in range(8):
                check_y = i + move_y[k]
                check_x = j + move_x[k]
                
                if 0 <= check_y < row and 0 <= check_x < col:
                    if game_map[check_y][check_x] == "*":
                        bomb_count += 1
                        
            print(bomb_count, end="")
    print()
```
