```python
# while 반복문 사용하기

# 초기식
# while 조건식:
#     반복할 코드
#     변화식

i = 0

while i < 10:
    print("hello world")
    i += 1   # i = i + 1

for _ in range(10):
    print("hello world!")
```

```python
# 초깃값을 1부터 시작하기

i = 1
while i <= 10:
    print("kello world", i)
    i +=1
```

```python
i = 10
while i > 0:
    print("hello world", i)
    i = i - 1
```

```python
# 입력한 횟수대로 반복하기

count = int(input("반복할 횟수 입력: "))

i = 0
while i < count:
    print("Hello World", i)
    i += 1
```

```python
# 연습문제

i = 2
j = 5

while i <= 32 or j >= 1:
    print(i, j)
    i *= 2
    j -= 1
```
