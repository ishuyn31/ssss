```python
# 문자열 여러 줄을 파일에 쓰기, 읽기

with open('./data/hello.txt', "w") as file:
    for i in range(3):
        file.write(f"hello world {i}\n")
```

```python
lines = [
    "안녕하세요.\n",
    "파이썬\n",
    "코딩 도장입니다.\n"
]

with open("./data/hello.txt", "w") as f:
    f.writelines(lines)
```

```python
## 파일의 내용을 한 줄씩 리스트로 가져오기
with open("./data/hello.txt", "r") as file:
    lines = file.readlines()
    print(lines)
```

```python
## 파일의 내용을 한 줄씩 읽기
with open("./data/hello.txt", "r") as f:
    line = None
    while line != "":
        line = f.readline()
        print(line.strip("\n"))
```

```python
## for 반복문으로 파일의 내용을 줄 단위로 읽기
with open("./data/hello.txt", "r") as f:
    for line in f:
        print(line.strip('\n'))
```

```python
# 파이썬 객체를 파일에 저장하기, 가져오기

import pickle

name = "james"
age = 17
adress = '서울시 서초구 반포동'
scores = {
    'Korean': 90,
    'English': 95,
    'Mathematics': 85,
    'Science': 82
}

with open("./data/james.p", "wb") as file:
    pickle.dump(name, file)
    pickle.dump(age, file)
    pickle.dump(adress, file)
    pickle.dump(scores, file)
```

```python
import pickle

with open("./data/james.p", "rb") as file:
    name = pickle.load(file)
    age = pickle.load(file)
    adress = pickle.load(file)
    scores = pickle.load(file)

    print(name)
    print(age)
    print(adress)
    print(scores)
```
